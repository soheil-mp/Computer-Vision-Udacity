# Creating a Filter, Edge Detection

### Import resources and display image

```python
# Importing the libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the image
image = cv2.imread("./images/city_hall.jpg")

# Making a copy of image
image_copy = np.copy(image)

# Covert from BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Visualizing the image
plt.imshow(image_copy)
```

### Convert the image to grayscale

```python
# Convert to grayscale for filtering
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# Visualizing the image
plt.imshow(gray, cmap = 'gray')
```

### Create a custom kernel

```python
# 3x3 array for edge detection (detects vertical edges and ignore horizontal edges)
# Sobel filter 
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])

# Performing convolution
filtered_image = cv2.filter2D(gray, -1, kernel) # second argument is the type of output image. -1 means that the output image will have the same type as the input image

# Visualizing the image
plt.imshow(filtered_image, cmap = 'gray')
```

### Create binary image

```python
# threshold = 100

# Create a threshold which sets all values above that threshold to white or 255
retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)

# Visualizing the image
plt.imshow(binary_image, cmap = 'gray')
```

#### Lower threshold will cause detecting on more edges. in other word i causes more noise that sometimes we don't need them.

```python
# threshold = 50

# Create a threshold which sets all values above that threshold to white (255)
retval, binary_image = cv2.threshold(filtered_image, 50, 255, cv2.THRESH_BINARY)

# Visualizing the image
plt.imshow(binary_image, cmap = 'gray')
```

# Video codes

### Import resources and display image

```python
# Importing the libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Loading the image
image = cv2.imread("./images/sunflower.jpg")

# Making a copy of image
image_copy = np.copy(image)

# Converting BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Visualizing the image
plt.figure(figsize = (20, 10))
plt.imshow(image_copy)
```

```python
# Converting to the grayscale image
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# Visualizing the image
plt.figure(figsize = (20, 10))
plt.imshow(gray, cmap = 'gray')
```

### Implement Canny Edge Detection

```python
# Defining lower and upper threshold for hysteresis
lower = 120
upper = 240 # it is recommended to use a ration of 1:2 or 1:3 for better results

# Canny filter
edges = cv2.Canny(gray, lower, upper)

# Visualizing the image
plt.figure(figsize = (20, 10))
plt.imshow(edges, cmap = 'gray')
```

```python
# Trying canny with "wide" and "tight" threshold
wide = cv2.Canny(gray, 30, 100) # wide because it has a very low lower threshold
tight = cv2.Canny(gray, 180, 240) # tight since it has a narrower range for acceptable edges

# Visualizing the image
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 10))

ax1.imshow(wide, cmap = 'gray')
ax1.set_title("Wide")

ax2.imshow(tight, cmap = 'gray')
ax2.set_title("Tight")
```

# K-means Clustering

### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
## TODO: Check out the images directory to see other images you can work with
# And select one!
image = cv2.imread('images/monarch.jpg')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

### Prepare data for k-means

```python
# Reshape image into a 2D array of pixels and 3 color values (RGB)
pixel_vals = image.reshape((-1,3))

# Convert to float type
pixel_vals = np.float32(pixel_vals)
```

### Implement k-means clustering

```python
# define stopping criteria
# you can change the number of max iterations for faster convergence!
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)

## TODO: Select a value for k
# then perform k-means clustering
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# reshape data into the original image dimensions
segmented_image = segmented_data.reshape((image.shape))
labels_reshape = labels.reshape(image.shape[0], image.shape[1])

plt.imshow(segmented_image)
```

```python
## TODO: Visualize one segment, try to find which is the leaves, background, etc!
plt.imshow(labels_reshape==0, cmap='gray')
```

```python
# mask an image segment by cluster

cluster = 0 # the first cluster

masked_image = np.copy(image)
# turn the mask green!
masked_image[labels_reshape == cluster] = [0, 255, 0]

plt.imshow(masked_image)
```

# Codes

### Importing resources and display image

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
# Loading the image
image = cv2.imread("./images/monarch.jpg")

# Making a copy
image_copy = np.copy(image)

# Converting BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Visualizing the image
plt.imshow(image_copy)
```

### Prepare data for k-means

```python
print("Dimension of image before reshaping: ", image_copy.shape)

# Reshaping image into a 2D array of pixels and 3 color values (RGB)
pixel_vals = image_copy.reshape((-1, 3))
print("Dimension of image after reshaping: ", pixel_vals.shape)

# Conver to float numbers
pixel_vals = np.float32(pixel_vals)
```

### Implement k-means clustring

### K = 2

```python
# Define stopping criteria
criteria =  (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) # critera tells algorithms when to stop # arguments: 1- here we are making criteria of epsilon or a max number of iterations 2- max number of iterations which is 10 3- epsilon which if the clusters mover less than this after some iterations, the algorithm stops, it reached convergence.

# Perform k-means clustring
k = 2
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS) # arguments: 1- 2D array image 2- value of K or clusters 3- labels we want 4- stop criteria which we have to define 5- number of attempts 6- how we choose our initial center points which is randomely

# Convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# Reshape to original size of image_copy
segmented_image = segmented_data.reshape(image_copy.shape)
labels_reshape = labels.reshape(image_copy.shape[0], image_copy.shape[1])

# Visualizing the image
plt.imshow(segmented_image)
```

```python
# Visualizing one segment
print("==> Visualizing segments for leaves and background")
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 10))
ax1.imshow(labels_reshape == 0, cmap = 'gray')
ax1.set_title("Segment 1")
ax2.imshow(labels_reshape == 1, cmap = 'gray')
ax2.set_title("Segment 2")
```

```python
# Mask image segment
masked_image = np.copy(image_copy)
masked_image[labels_reshape == 1] = [0, 0, 0]

# Visualizing the image
plt.imshow(masked_image)
```

```python
# Mask image segment
masked_image = np.copy(image_copy)
masked_image[labels_reshape == 0] = [0, 0, 0]

# Visualizing the image
plt.imshow(masked_image)
```

```python
# Size of images
plt.figure(figsize = (20, 10))

# Making mask for each segment and visualizing it
for i in range(0, 2):
    masked_image = np.copy(image_copy)
    masked_image[labels_reshape == i] = [255, 0, 0]
    ax = plt.subplot(1, 2, i + 1)
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.set_title("segment " + str(i + 1))
    ax.imshow(masked_image)
  
# Eliminating space between images
plt.subplots_adjust(wspace = 0, hspace = 0)

plt.show()
```

### K = 6

```python
# Loading the image
image = cv2.imread("./images/monarch.jpg")

# Making a copy
image_copy = np.copy(image)

# Converting BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

# Visualizing the image
plt.imshow(image_copy)
```

```python
# Define stopping criteria
criteria =  (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) # critera tells algorithms when to stop # arguments: 1- here we are making criteria of epsilon or a max number of iterations 2- max number of iterations which is 10 3- epsilon which if the clusters mover less than this after some iterations, the algorithm stops, it reached convergence.

# Perform k-means clustring
k = 6
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS) # arguments: 1- 2D array image 2- value of K or clusters 3- labels we want 4- stop criteria which we have to define 5- number of attempts 6- how we choose our initial center points which is randomely

# Convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# Reshape to original size of image_copy
segmented_image = segmented_data.reshape(image_copy.shape)
labels_reshape = labels.reshape(image_copy.shape[0], image_copy.shape[1])

# Visualizing the image
plt.imshow(segmented_image)
```

```python
# Visualizing one segments
print("==> Visualizing segments for leaves and background")
f, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(nrows = 1, ncols = 6, figsize = (20, 20))

# Visualizing the image
ax1.imshow(labels_reshape == 0, cmap = 'gray')
ax2.imshow(labels_reshape == 1, cmap = 'gray')
ax3.imshow(labels_reshape == 2, cmap = 'gray')
ax4.imshow(labels_reshape == 3, cmap = 'gray')
ax5.imshow(labels_reshape == 4, cmap = 'gray')
ax6.imshow(labels_reshape == 5, cmap = 'gray')

# Adding title
for index, i in enumerate((ax1, ax2, ax3, ax4, ax5, ax6)):
    i.set_title("segment " + str(index + 1))
    
# Eliminating the axis 
for i in (ax1, ax2, ax3, ax4, ax5, ax6):
    i.xaxis.set_visible(False)
    i.yaxis.set_visible(False)
    
# Eliminating space between images
plt.subplots_adjust(wspace = 0, hspace = 0)
```

```python
# Mask image segment
masked_image = np.copy(image_copy)

plt.figure(figsize = (20, 10))
# Making mask for each segment and visualizing it
for i in range(0, 6):
    masked_image = np.copy(image_copy)
    masked_image[labels_reshape == i] = [255, 0, 0]
    ax = plt.subplot(1, 6, i + 1)
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    ax.set_title("segment " + str(i + 1))
    ax.imshow(masked_image)
  
# Eliminating space between images
plt.subplots_adjust(wspace = 0, hspace = 0)

plt.show()
```
